```python
df = df.rename(columns={
    'vhigh': 'harga',
    'vhigh.1': 'biaya_pemeliharaan',
    '2': 'pintu',
    '2.1': 'kapasitas',
    'small': 'ukuran',
    'med': 'rating_keselamatan',
    'unacc': 'evaluasi'
})
df
```

```python
import pandas as pd

# Mengubah nilai-nilai dalam kolom menjadi nilai numerik
df['harga'] = df['harga'].map({"low": 0, "med": 1, "high": 2, "vhigh": 3})
df['biaya_pemeliharaan'] = df['biaya_pemeliharaan'].map({"low": 0, "med": 1, "high": 2, "vhigh": 3})
# df['pintu'] = df['pintu'].map({"small": 0, "med": 1, "big": 2})
# df['kapasitas'] = df['kapasitas'].map({"low": 0, "med": 1, "high": 2})
df['ukuran'] = df['ukuran'].map({"small": 0, "med": 1, "big": 2})
df['rating_keselamatan'] = df['rating_keselamatan'].map({"low": 0, "med": 1, "high": 2})
df['evaluasi'] = df['evaluasi'].map({"unacc": 0, "acc": 1, "good": 2, "vgood": 3})
```

```python
df.info()
```

```python
from sklearn.cluster import KMeans

clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(df)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Menghitung inersia untuk setiap jumlah cluster dari 1 hingga 10
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df[['kapasitas', 'rating_keselamatan']])  # Menggunakan contoh fitur 'kapasitas' dan 'rating_keselamatan'
    inertia.append(kmeans.inertia_)

# Buat plot Elbow
plt.figure(figsize=(8, 4))
sns.lineplot(x=range(1, 11), y=inertia)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# Buat Object KMeans
km5 = KMeans(n_clusters=4).fit(df[['kapasitas', 'rating_keselamatan']])

# Menambahkan Kolom Label Pada Dataset
df['Labels'] = km5.labels_

# Buat Plot Kmeans
plt.figure(figsize=(8,4))
sns.scatterplot(x=df['kapasitas'], y=df['rating_keselamatan'], hue=df['Labels'], palette=sns.color_palette('hls', 5))

plt.title('KMeans dengan 4 Cluster')
plt.xlabel('Kapasitas Penumpang')
plt.ylabel('Rating Keselamatan')
plt.show()
```
